```python
import numpy as np # linear algebra, used for numeric evaluations
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv), used for analysing data
import seaborn as sb  #Visualization libraries (This is for pair plot)
import matplotlib.pyplot as plt

#from jupyterthemes import jtplot
#jtplot.style(theme ="monaki", context = "notebook", ticks =True, grid = False)

#Read the file from 
students = pd.read_csv("Admissions.csv")
#Print the head
students.head()
students
#The output is a regression model showing what chances does the student carries
#To drop a column, students.drop("NAme Column", axis = '1', inplace = True).  ->Inplace for releasing memory

#Check for missing data 
students.isnull().sum()
#To get the information of the csv files with formats
students
```

```python
#Checking what we have in the data
students.columns
#Create labels
#X is the independent variable. 
#Y is a dependent variable (Target Variable) : Chance of Admission
#Drop the columns that are not required 
X = students.drop(columns = ['Chance'])
Y = students['Chance']

#Find the shapes of X , Y -> Dimensions
X.shape
Y.shape

#Fit them into numpy array
X = np.array(X)
Y = np.array(Y)

#Reshape Y to use further algorithm (Normalize)
Y = Y.reshape(-1,1)
Y.shape
```

```python
#SkLearn is used to Scale data to avoid biasing
from sklearn.preprocessing import StandardScaler ,MinMaxScaler

#Using a standard scaler form to scale the data
scaler_x = StandardScaler()
X = scaler_x.fit_transform(X)

scaler_y = StandardScaler()
Y = scaler_y.fit_transform(Y)
#The warning conversion if for 2 to 2.0
```

```python
#Split for training and testing
from sklearn.model_selection  import train_test_split
#Initialize the split parts, allocate 25% of data to testing
x_test, x_train, y_test, y_train = train_test_split(X, Y, test_size = 0.25)
```

```python
#Import the Lin Reg and error/cost function
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

#Create object
LinearReg = LinearRegression()
#Fit data in the simple linear regression model
LinearReg.fit(x_train, y_train)

#Find the accuracy over test data 
accuracy_model = LinearReg.score(x_test, y_test)
accuracy_model
```

```python
#Plot this 
y_predict = LinearReg.predict(x_test)
plt.plot(y_test, y_predict, '^', color = 'r')
#This graph is in the transformed units (Scaled Version between (-1,1))
```

```python
#Invert the scaling matrix
y_predict_original = scaler_y.inverse_transform(y_predict)
y_test_original = scaler_y.inverse_transform(y_test)
#Plot this now
plt.plot(y_test_original, y_predict_original, '^')
plt.xlabel('Original Data')
plt.ylabel('Predicted Data')
plt.show()
```

```python
#The abovel model uses Regression In Multilple Variables to predict the chances of admission
#Now we will create a Artificial Neural Network
```

```python
#Imports
#Using TensorFlow as backend (A Google Framework to train Neural Networks)
import sys
sys.executable
import tensorflow as tf
from tensorflow import keras
#Dene -> To build dense artificial network
#Activation -> Activation Functions like Rectified Linear Units (RelU)
#Dropout-> Regularization Technique to avoid over fitting. To mak eusre the neurons do not drop any dependency among each other
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
#Create the fist layer with 50 neuons and 2 input features, and then add the activation fuctions
ANN_model = keras.Sequential()
ANN_model.add(Dense( 50, input_dim =2))
ANN_model.add(Activation('relu'))



#Create two more layers with about 20 neurons, and drop 40% of the previous ones
ANN_model.add(Dense(20))
ANN_model.add(Activation('relu'))
ANN_model.add(Dropout(0.40))

ANN_model.add(Dense(10))
ANN_model.add(Activation('relu'))
ANN_model.add(Dropout(0.4))

#Converge the neural network model
#ANN_model.add(Dense(50))
#ANN_model.add(Activation('Linear'))

#ANN_model.add(Dense(15))
#ANN_model.add(Activation('Linear'))

#Finally converges to the result
ANN_model.add(Dense(1))
```

```python
#Compile the neural netwroking model, mse -> mean squared error
ANN_model.compile(loss = 'mean_squared_error', optimizer = 'Adam')

#Iteration
epoch_history = ANN_model.fit(x_train, y_train, epochs = 100, batch_size = 10)

#Get the result
result = ANN_model.evaluate(x_test, y_test)
accuracy_ANN = 1- result
print("Accuracy : {}".format(accuracy_ANN))
```

```python
#Plot the epoch history
epoch_history.history.keys()
#Plot withplt
plt.plot(epoch_history.history['loss'])
plt.title("Model Loss during the training")
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error/Training Loss')
plt.legend(['Training Loss'])
```

```python
#KPIs
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt

#Number of features
k = x_test.shape[1]
#No. of data samples
n = len(x_test)

RMSE = float(format(np.sqrt(mean_squared_error(y_test_original, y_predict_original)), '.3f'))
MSE = mean_squared_error(y_test_original, y_predict_original)
MAE = mean_absolute_error(y_test_original, y_predict_original)
r = r2_score(y_test_original, y_predict_original)
adjR2 = 1- ((1-r)*(n-1))/(n-k-1)

print("RMSE: ", RMSE)
print("MSE: ", MSE)
print("MAE: ", MAE)
print("R2_Score: ", r)
print("Adjusted R2: ", adjR2)


```
